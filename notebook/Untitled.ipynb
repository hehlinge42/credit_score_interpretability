```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/input.csv", sep=";")
data
```

```python
plt.figure(figsize=(15, 10))
saving_dict = {}
for saving in ["A43", "A46", "A42", "A40", "A41", "A49", "A44", "A45", "A410", "A48"]:
    data_saving = data[data["Purpose"] == saving]
    data_saving = data_saving["CreditAmount"]
    data_saving.plot(kind='density', label=saving)
    #saving_dict[saving] = data_saving["CreditAmount"].tolist()
plt.legend()
```

```python
plt.figure(figsize=(15, 10))
for k, v in saving_dict.items():
    plt.hist(v, label=k, alpha=0.5)
plt.legend()
```
